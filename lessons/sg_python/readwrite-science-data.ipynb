## 2. Starting simple: CSV and copy paste


```python
# Basic CSV with pandas

import pandas

# Modify path as needed
table1 = pandas.read_csv('./table1',
                         sep='\t',index_col=0)

# Q: What are the other arguments besides the path? Specifically what is  " \t "  ;   what would " \s  " do?

table1
# Note that the column names are automatically imported, since we specified an index column

# Some info: Series, DataFrames, Panels
```

```python
# Accessing

# Here we create some spectral mixtures, stripping out the pixel x and y location
mix_50_50 = table1.T.ix[2:,'Capoeira']*.5 + table1.T.ix[2:,'Photo Shade']*.5
mix_75_25 = table1.T.ix[2:,'Capoeira']*.75 + table1.T.ix[2:,'Photo Shade']*.25
mix_25_75 = table1.T.ix[2:,'Capoeira']*.25 + table1.T.ix[2:,'Photo Shade']*.75

# Q: What does 'tabel.T' do? Why do we use it?

mix_50_50
# The index carries over... we can label the mixes and plot...
```

```python
# Plotting: Dumb plots

plot(mix_50_50)
```

```python
# Plotting: Auto labels using pandas

mix_50_50.plot()
```

```python
# Q: Plot all the landcover types, labled, and without the x y columns

# Hint: Remember the mixture syntax? Try modifying the expression
mix_50_50 = table1.T.ix[2:,'Capoeira']*.5 + table1.T.ix[2:,'Photo Shade']*.5
```

```python
# Creating a new DataFrame

# Passing in series names for our series
spectral_mix = pandas.DataFrame({'75% Capoeira, 25% Shade' : mix_75_25, 
                                 '50% Capoeira, 50% Shade' : mix_50_50, 
                                 '25% Capoeira, 75% Shade': mix_25_75})
# To plot:
figure()
spectral_mix.plot()
legend(loc=2)
xlabel('Wavelength (nm)')
ylabel('% Reflectance')
# Note the legend carries from the labels, and the x-axis
```

```python
# Working with images

# Not needed if notebook is started with --pylab
from pylab import *

#Make plots interactive
ion()

#Importing our data-- change path as needed
mat = scipy.io.loadmat('./etopo5_matlab.mat')
#Python imports the .mat file as a dictionary, we need to extract our array
etopo_data = mat['etopo5_data']
#convert to double float
etopo_data = double(etopo_data)
#equivlent to imagesc in matplab
#note that axis equal tight is the default in python
imshow(etopo_data)
colorbar()
title('ETOPO5 Topographic and Bathymetric data')
```

```python
#Scaleing data:
#Requires using the vmin and vmax args
figure()
imshow(etopo_data, vmin=0, vmax=5000)
title('ETOPO5 Topographic and Bathymetric data')
#note-- easiest to call colorbar after scaling
colorbar()
```
